# Sample URL and analyze

```python
import pandas as pd
all_urls = pd.read_csv("lotsawa_urls.csv")
print(len(all_urls['url']), len(set(all_urls['url'])))
all_urls = all_urls.groupby("url").aggregate(lambda x: '\n'.join(x)).reset_index()
```

```python
all_urls[:10]
```

```python
from lotsawa_util import *
all_pairs = []
for i, row in all_urls.iterrows():
    url = row["url"]
    print("processing...", i, url)
    pairs = process_one_url(url)
    all_pairs += [(pair[0], pair[1], pair[2], url, row['topic_page']) for pair in pairs]
```

```python
# Error prone extraction class
wrong_ec = {"TibetanVerse|Explanation","TibetanVerse|Heading3","TibetanExplanation|EnglishText"}
all_pairs = [all_pairs[i] for i in range(len(all_pairs)) if all_pairs[i][2] not in wrong_ec]
output_interlace = pd.DataFrame(all_pairs, columns=[["source", "target", "extraction_class", "url", "topics"]])
```

```python
output_interlace.to_csv("bisent_lotsawa_interlace.csv", index=False)
```

## estimate headroom

```python
chunk_text_urls = list(set(all_urls["url"]) - set(data["url"]))
print("#urls that doesn't contain interlace bo-en text: ", len(chunk_text_urls))
```

```python
import random
random.shuffle(chunk_text_urls)
chunk_text_url_sample = chunk_text_urls[:10]
chunk_text_url_sample
```

```python
total_shay = 0
for url in chunk_text_url_sample:
    tmp = url.split("/")
    bo_url = "/".join(tmp[:3] + ["bo"] + tmp[3:])
    r = requests.get(bo_url)
    soup = BeautifulSoup(r.content, 'html.parser')
    bo_text = soup.select("#maintext")[0].text
    n_shay = bo_text.count("།")
    total_shay += n_shay
    print("# of shay's: ", n_shay, " at ", bo_url)
```

```python
total_shay * 348 / 10
```

```python
set([t['class'][0] for t in soup.select("#maintext > p") if 'class' in t.attrs])
```

```python
pairs = []
opening = ""
startings = {'TibetanExplanation', 'HeadingTib', 'TibetanVerse'}
endings = {'Explanation', 'Heading3', 'EnglishText'}
for t in soup.select("#maintext > p"):
    if 'class' not in t.attrs:
        if t.text.strip(): print("No class??", t)
    else:
        if t['class'][0] in endings and opening:
            pairs.append((opening, t.text))
            opening = ""
        elif t['class'][0] in startings:
            opening = t.text
```
